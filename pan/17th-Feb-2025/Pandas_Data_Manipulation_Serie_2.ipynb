**Combining Filters**
We can combine conditional statements using the bitwise operators for:

* and: &
* or: |

What if we wanted to filter for the baked goods that were baked by Jade and had a price > $5?

First, we can create a filter for Jade.

```python
df[(df['Student'] == 'Jade') & (df['Price']>5)]['Item']
```

```python
# Create a filter for Jade
filter_jade = df['Student']=='Jade'
filter_jade
```

```python
# Create a filter for Price
filter_price = df['Price'] > 5
filter_price
```

To combine the filters, we just use the correct operator for the logic we want to apply. Since we want rows that meet both conditions, we will use the bitwise and operator (& ). It is important to note that combining filters simply produces a new filter with a single True/False for every row.

```python
# Combine the filters
filters_combined = filter_jade & filter_price
filters_combined
```

We can then use the combined filter to slice out the rows that meet both conditions.

```python
# Apply the combined filter
df.loc[filters_combined]
```

```python
df[(df['Student']=="Jade") & (df['Price']>5)]
```

**No matches**

It is quite possible that there no values will meet the criteria for a filter. For example, in our data, there are no items that are less than $2 and made by Jade. Let's see what happens when we attempt to apply the filter when no values will match.

```python
# Create a new filter for less than 2
filter_price_less = df['Price'] < 2
# Combine the filters
df.loc[ filter_price_less & filter_jade]
```

**Code Efficiency vs Readability**

As your coding skills grow, you will start to recognize ways in which you could reduce the number of lines of code required for a particular operation.

If you wanted to filter the dataframe on multiple conditions without saving the filters as a variable, you can do so as long as you put each filter in a tuple surrounded by parenthesis.

```python
# Alternate method for combing filters without saving them first
df.loc[(df['Student']=='Jade') & (df['Price'] > 5)]
df[()&()]
```

You should be careful about how much you try to accomplish in one line of code, though. There is a tradeoff between code efficiency and readability.

**Using df.groupby to get grouped statistics**

To answer the next two stakeholder questions, we will use a special dataframe method called groupby.

Pandas makes it very easy to calculate aggregations grouped by one or more columns.

To use groupby:

* start with a dataframe (not a Series/single column)
* followed by groupby()with the names of the column(s) included in the parenthesis
* an aggregation method such as .sum()

4) Which student sold the largest quantity of items?

To address this question we will group our data by Student. We use .sum() as the aggregation calculation.

```python
df.groupby('Student').sum()['Cumulative Quantity (Day 4)']
```

```python
# groupby needs an aggregation functon
student_sums = df.groupby("Student").sum()
student_sums
```

In the result above, we can see that:

* The column we used in our Groupby method is the index. (In this case, "Student")
* We have 1 row per group (the Students).
* Each numeric column is present.

Since we are specifically interested in the Cumulative Quantity by Sudent, we can slice out the 'Cumulative Quantity (Day 4)' before the .sum() to reduce unnecessary output.

```python
# Obtain the cumulative profit per student
quantity_by_student = df.groupby('Student')['Cumulative Quantity (Day 4)'].sum()
# Sort from highest to lowest profit
quantity_by_student.sort_values(ascending=False)
```

As we can see, Sally has sold the largest quantity of baked goods (153 items), followed by Hugo (150 items).

Renaming columns can make your life easier. Deciding which columns to rename and how to rename them is subjective, but we have selected some very abbreviated column names or ambiguous columns to rename for easier interpretation:

```python
# dictionary substitution using rename method
df = df.rename(rename_dict,axis=1)
df.head()
```

We don't want to have to look row by row, so let's use .sum() to add up all of the True values. When using .sum(), a True value will be counted as 1. A False value will not be counted (or counted as 0).
